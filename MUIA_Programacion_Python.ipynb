````python
# slicing unidimensional arrays
array = np.arange(20)**2
print(array)
# important to understand the concept of deep and shallow copy
# this is deep copy - you break with the original
# you can read https://docs.python.org/3/library/copy.html
arrayoriginal = array.copy()
#pay attention at this next line - you add array[2] to each an every value from positions 3,4,5 and 6
print(array[2]+array[4])
#pay attention at this next line - you add array[2] to each an every value from positions 3,4,5 and 6
print(array[2]+array[3:7])
print(array[2:7])

# from start to position 8, exclusive, set every 2nd element to 100
# equivalent to array[0:8:2] = 100;
# this is shallow copy, my friends
# slices [1:] and [:-1] mean all but the first and all but the last elements of the array
array2=array
array3=array
array2[:8:2] = 100
# this is another way to say the same thing
array3[0:8:2] = 100
print(array2)
print(array3)
#this is reversed
print(arrayoriginal[::-1])
# you can also do loops
for i in arrayoriginal:

    print(i**(1 / 0.5))
````

````python
# now let us explain something similar in multidimensional matrixes
def alinearf(x, y):

    return 10 * x + y

# and now look at how we declare a matrix
anarray = np.fromfunction(alinearf, (4, 3), dtype=int)

# print the entire array
print(anarray)
# print an element
print(anarray[2,2])
# print each row in the second column of anarray
print(anarray[0:4,1])
# a similar result
print(anarray[:, 1])
````

````python
# look at how flat can help
# example of the use of loops
for element in anarray.flat:

    print(element)
    element = element + 2
    print(element)
# you can add 2 to each element of the matrix
anarraynew = anarray + 2
# you can see the result in the array
print(anarraynew)
print(anarray)
# but look at how the original array is influenced
anarray +=2
# the syntax above modifies the initial array
print(anarray)
````

````python
# stacking arrays
# we create 2 arrays
array1 = np.floor(10 * randomg.random((3, 4)))
print(array1)
array2 = np.floor(10 * randomg.random((3, 4)))
print(array2)
# we stack them vertically
# attention at the double parenthesis
# vertically
array3v = np.vstack((array1, array2))
print(array3v)
# horizontally
array3h = np.hstack((array1, array2))
print(array3h)
# in columns for 1-D arrays
xarray1c = np.floor(10 * randomg.random((1, 2)))
xarray2c = np.floor(10 * randomg.random((1, 2)))
# stack arrays in sequence vertically (row wise).
xarray3c = np.row_stack((xarray1c, xarray2c))
#2-D arrays are stacked as-is,
xarray3r = np.column_stack((xarray1c, xarray2c))
xarray3d = np.hstack((xarray1c, xarray2c))
xarray3e = np.vstack((xarray1c, xarray2c))
print(xarray1c)
print(xarray2c)
print(xarray3c)
print("column_stack result:")
print(xarray3r)
print("hstack result:")
print(xarray3d)
print("vstack result:")
print(xarray3e)
````

````python
# for instance, in fluid dynamics, you have the Airy functions
# more info here on these functions https://en.wikipedia.org/wiki/Airy_function
# how to use the function in Python here https://docs.scipy.org/doc/scipy/reference/generated/scipy.special.airy.html#scipy.special.airy
import numpy as np
from scipy import special
# remember, linspace from -25 to 15 and 301 elements
x = np.linspace(-25, 15, 301)
ai, aip, bi, bip = special.airy(x)
````

# **5**. Tensorflow

````python
# Information from https://www.tensorflow.org/overview
# and from https://en.wikipedia.org/wiki/TensorFlow
# more specifically, from https://www.tensorflow.org/resources/learn-ml/basics-of-machine-learning
# and finally this one https://www.tensorflow.org/tutorials
# Tensorflow is an open-source for machine learning and AI with a focus on
# training and inference of deep neural networks
# what is a neural network?
# neural networks consists of neurons, synapses, weights, biases, and functions
# what is a deep neural network?
# it is an artificial neural network with multiple layers
# between input and output layers
# let us look at the first piece of code using tensorflow
````

````python
import tensorflow as tf
print("TensorFlow version:", tf.__version__)
````

````python
# we use the NIST database of handwritting characters
# reference https://en.wikipedia.org/wiki/MNIST_database
mnist = tf.keras.datasets.mnist
# we load the data with this instruction
(x_train, y_train), (x_test, y_test) = mnist.load_data()
# normalisation: scale these values to a range of 0 to 1 by dividing the values by 255.0
# this instruction also converts the sample data from integers to floating-point numbers
x_train, x_test = x_train / 255.0, x_test / 255.0
````

````python
# we use a specific model
# sequential is a layer stacking model
# the layers used here are flatten, dense and dropout
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(28, 28)),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(10)
])
````

````python
# slices [1:] and [:-1] mean all but the first and all but the last elements of the array
predictions = model(x_train[:1]).numpy()
predictions
````

````python
#tf.nn.softmax function converts these logits to probabilities
tf.nn.softmax(predictions).numpy()
````

````python
# it is also usual to create a loss function
loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
# this is the type of object in loss_fn
print(loss_fn)
````

````python
# now we compile the model
model.compile(optimizer='adam',
              loss=loss_fn,
              metrics=['accuracy'])
# and we fit the model using the training data
model.fit(x_train, y_train, epochs=5)
````

````python
# we can evaluate the model
model.evaluate(x_test,  y_test, verbose=2)
# the accuracy is the percentage the model has been trained on this dataset
````

````python
# we can also return a probability
probability_model = tf.keras.Sequential([
  model,
  tf.keras.layers.Softmax()
])
probability_model(x_test[:5])
````
